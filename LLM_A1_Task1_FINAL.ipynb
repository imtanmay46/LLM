***LLAMA3***

```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

tokenizer_llama = AutoTokenizer.from_pretrained("meta-llama/Meta-Llama-3-8B-Instruct", use_auth_token=True)
model_llama = AutoModelForCausalLM.from_pretrained(
    "meta-llama/Meta-Llama-3-8B-Instruct",
    device_map='auto',
    load_in_4bit=True
)
```

```python
def generate_response(prompt, model, tokenizer):
    inputs = tokenizer(prompt, return_tensors="pt")
    outputs = model.generate(**inputs, max_length=150)
    response = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return response
```

**FACTUAL HALLUCINATIONS FOR LLAMA3**

*EXAMPLE-1*

```python
prompt1 = "Give only one name of the country which won the ODI Men's Cricket World Cup in 2018?"
response1 = generate_response(prompt1, model_llama, tokenizer_llama)

print(f"\n\nPrompt: {prompt1}\n")
print(f"LLAMA Response: {response1}")
```

*EXAMPLE-2*

```python
prompt2 = "Did Charles Leclerc win the 2020 Italian Grand Prix?"
response2 = generate_response(prompt2, model_llama, tokenizer_llama)

print(f"\n\nPrompt: {prompt2}\n")
print(f"LLAMA Response: {response2}")
```

*EXAMPLE-3*

```python
prompt3 = "Is Lewis Hamilton an 8-time World Driver Champion? Yes or No?"
response3 = generate_response(prompt3, model_llama, tokenizer_llama)

print(f"\n\nPrompt: {prompt3}\n")
print(f"LLAMA Response: {response3}")
```
