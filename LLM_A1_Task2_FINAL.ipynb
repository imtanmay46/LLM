```python
sentence_model_llama3 = SentenceTransformer('all-MiniLM-L6-v2')
embeddings_llama3 = sentence_model_llama3.encode(knowledge_data_llama3)
```

```python
sentence_model_openhathi = SentenceTransformer('paraphrase-multilingual-MiniLM-L12-v2')
embeddings_openhathi = sentence_model_openhathi.encode(knowledge_data_llama3)
```

```python
for idx, (text, embedding) in enumerate(zip(knowledge_data_llama3, embeddings_llama3)):
    collection_llama3.add(
        ids=[f"doc_{idx}"],
        documents=[text],
        embeddings=[embedding.tolist()]
    )
```

```python
for idx, (text, embedding) in enumerate(zip(knowledge_data_openhathi, embeddings_openhathi)):
    collection_openhathi.add(
        ids=[f"doc_{idx}"],
        documents=[text],
        embeddings=[embedding.tolist()]
    )
```

```python
def retrieve_relevant_docs(sentence_model, collection, query, top_k=3):
    query_embedding = sentence_model.encode([query])[0]
    results = collection.query(
        query_embeddings=[query_embedding.tolist()],
        n_results=top_k
    )
    return results['documents']
```

```python
def generate_response(query, model, tokenizer, sentence_model, collection, top_k=2, max_new_tokens=100):
    relevant_docs = retrieve_relevant_docs(sentence_model, collection, query, top_k=top_k)

    augmented_query = query + "\nRelevant Information: " + "\n# ".join(relevant_docs[0]) + "\n\n###\n\n"

    inputs = tokenizer(augmented_query, return_tensors="pt")
    output = model.generate(**inputs, max_new_tokens=max_new_tokens)
    response = tokenizer.decode(output[0], skip_special_tokens=True)

    return response
```
